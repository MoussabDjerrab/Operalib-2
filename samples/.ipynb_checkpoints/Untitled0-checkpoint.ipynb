```python
from operalib import FISTA
from operalib import LearningRate
from operalib import Risk
from operalib import Kernels
from operalib import Model

import numpy as np
import scipy.io as io
```

```python
data = io.loadmat( 'DREAM3_size10Ecoli1.mat' )
time_serie = data[ 'data' ][ 0 ][ 0 ]
X = time_serie[ :-1, :]
y = time_serie[ 1:, :]

# A = np.random.rand( 10, 10 )
# A = np.dot( A.T, A )
# A = A / np.linalg.norm( A, 2 )
A = np.eye( 10 )
```

```python
est = Model.Model( Kernels.DecomposableKernel( 1, 10, A ), X )
rsk = Risk.OKVAR( 1e-3, 1e-3, 1e-3, 1, 1, 1 )

L = np.linalg.norm( np.dot( est.Gram, est.Gram ), 2 )
opt = FISTA.FISTA( rsk, L, 10, 100 )
```

```python
opt.fit( est, X, y )
```

```python
np.mean( ( est( X ) - y ) ** 2 )
```
